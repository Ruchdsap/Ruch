```python
import os 
import pandas as pd
import nltk
import numpy as np
```

```python
os.getcwd()
```

```python
fname = 'C:\\Users\\annam\\Downloads\\sentiment labelled sentences\\sentiment labelled sentences\\imdb_labelled.txt'
```

```python
lines = [l.strip() for l in open(fname).readlines()]
```

```python
len(lines)
```

```python
dataset = [l.split('\t') for l in lines]
```

```python
dataset[:10]
```

```python
df = pd.DataFrame(dataset, columns=['sentence','sentiment'])
```

```python
df
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokenized_sents = df['sentence'].apply(lambda x: word_tokenize(x))
```

```python
df['tokenized'] = tokenized_sents
```

```python
vocab = set()
for sent_toks in tokenized_sents:
    for t in sent_toks:
        vocab.add(t)
        
len(vocab)
```

```python
vocab = {w:i for i,w in enumerate(sorted(vocab))}
```

```python
vocab
```

```python
rand_wv = np.random.rand(len(vocab),300)
rand_wv.shape
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2,random_state=dom_state=42)
```

```python
train.head()
```

```python
X_train = []
for tok_sent in train['tokenized']:
    doc_vec = np.zeros(300)
    for t in tok_sent:
        word_index = vocab[t]
        word_vec = rand_wv[word_index]
        doc_vec += word_vec
    doc_vec = doc_vec/len(tok_sent)
    X_train.append(doc_vec)
    
X_train = np.array(X_train)
y_train = train['sentiment']
X_train.shape
```

```python
X_test = []
for tok_sent in test['tokenized']:
    doc_vec = np.zeros(300)
    for t in tok_sent:
        word_index = vocab[t]
        word_vec = rand_wv[word_index]
        doc_vec += word_vec
    doc_vec = doc_vec/len(tok_sent)
    X_test.append(doc_vec)
    
X_test = np.array(X_test)
y_test = test['sentiment']
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_true=y_test, y_pred=y_pred)
```
